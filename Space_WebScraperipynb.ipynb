```
# define a function to process the page
#iterate over all stories.

raw_stories = soup.select('.content')

def process_page(soup, stories):
    for story in raw_stories:
        headline = story.select_one('.article-name').get_text() # extract the headline
        author = story.select_one('.by-author > span').get_text().lstrip('\n') # extract the au name only
        synopsis = story.select_one('.synopsis').get_text().lstrip('\n').rstrip('\n') # extract synopsis get rid of extra. \n's
        date_time = story.select_one('.published-date').attrs['datetime']  #get datetime
        story ={'headline' : headline, 'author' : author, 'synopsis': synopsis, 'date_time': date_time}
        stories.append(story) # add dictionary to list

    return stories
    
    
stories = []
    
    
process_page(soup, stories)


#<time class="published-date relative-date" datetime="2019-07-29T19:00:00Z" 
#data-published-date="2019-07-29T19:00:00Z">2 hours ago</time>

#HOW TO GET RID OF THE '\n' in the output? .rstrip('\n')????
#headline = story.select_one('.article-name').get_text()
#author = story.select_one('.byauthor > span'.lstrip('\n')) #extract au name and strips the 2nd span tag
#synopsis = story.select_one(.synopsis)
#date_time = story_select_one('time').attrs(published-date.relative-date) ?? nope
#story ={'headline' : headline, 'author' : author, 'synopsis': synopsis, 
#'date_time': date_time}
```

```
# import all dependencies at the top
from time import time
from time import sleep
from random import randint
from IPython.core.display import clear_output
from warnings import warn
from bs4 import BeautifulSoup
import requests


#add function from above
# define a function to process the page
#iterate over all stories.

raw_stories = soup.select('.content')

def process_page(soup, stories):
    for story in raw_stories:
        headline = story.select_one('.article-name').get_text() # extract the headline
        author = story.select_one('.by-author > span').get_text().lstrip('\n') # extract the au name only
        synopsis = story.select_one('.synopsis').get_text().lstrip('\n').rstrip('\n') # extract synopsis get rid of extra. \n's
        date_time = story.select_one('.published-date').attrs['datetime']  #get datetime
        story ={'headline' : headline, 'author' : author, 'synopsis': synopsis, 'date_time': date_time}
        stories.append(story) # add dictionary to list

    #return stories
    return stories
    
    
stories = []
    
    
process_page(soup, stories)
    

  
# prepare for the monitoring logic
start_time = time() # note the system time when the program starts
request_count = 0 # track the number of requests made

# create a list to store the data in
stories = []

# variables to handle the request loop
has_next_page = True
MAX_REQUESTS = 100 # do not request more than 100 pages
page_number = 1
query = {'sort':'i', 'pg': page_number}
url = 'https://www.space.com/news'
headers = {'user-agent': 'webscraper - school project (socgrrrl@gmail.com)'}

while has_next_page and request_count <= MAX_REQUESTS:
  # keep the output clear
  clear_output(wait = True)
  
    # make an initial request
  response = requests.get(url, params=query, headers=headers)

  # make sure we got a valid response
  if(response.ok):
    # get the full data from the response
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    process_page(soup, stories)

    # check for the next page
    # look for the presence of element with class "listings-pagination-button listings-next "

    next_button = soup.select('.listings-pagination-button')
    has_next_page = len(next_button) > 0
    
  else:
    # display a warning if there are any problems
    warn('Request #: {}, Failed with status code: {}'.format(request_count, response.status_code))
  
  request_count += 1
  
  # go to sleep for a bit
  # we use a random number between 1 and 5 so
  # We can wait as long as 5 seconds to make a second request
  
  sleep(randint(1,3))
  
  # output some logs for monitoring
  elapsed_time = time() - start_time
  print('Requests: {}, Frequency: {} requests/s, {} articles processed.'.format(request_count, request_count/elapsed_time, len(stories)))
  
  # prepare for next iteration
  page_number += 1
      
print('Sraping complete')
print('Requests: {}, Frequency: {} requests/s, {} articles processed.'.format(request_count, request_count/elapsed_time, len(stories)))
```
